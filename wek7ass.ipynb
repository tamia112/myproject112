```python


#  fill missing values
df = df.fillna(0)
```

```python
#check mean values by diabetes status
grouped = df.groupby('Diabetes_binary').mean()
print(grouped)
```

```python
# Visualizations
sns.set_theme(style="whitegrid")
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(x=df.index, y="BMI", color="blue", linewidth=2)
plt.title("📈 Line Chart of BMI Across Patients", fontsize=16, fontweight="bold")
plt.xlabel("Index (Patient ID)")
plt.ylabel("BMI")
plt.show()
```

```python
#  Line Chart — BMI across Index
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y="BMI", color="blue", linewidth=2)
plt.title(" Line Chart of BMI Across Patients", fontsize=16, fontweight="bold")
plt.xlabel("Index (Patient ID)")
plt.ylabel("BMI")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x="Diabetes_binary", y="BMI", palette="Set2")
plt.title("🍏 Average BMI by Diabetes Status", fontsize=16, fontweight="bold")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Average BMI")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(data=df, x="Diabetes_binary", y="BMI", palette="Set2", hue="Diabetes_binary", legend=False)
plt.title("🍏 Average BMI by Diabetes Status", fontsize=16, fontweight="bold", fontname="Segoe UI Emoji")

plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Average BMI")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(
    data=df,
    x="Diabetes_binary",
    y="BMI",
    hue="Diabetes_binary",   
    palette="Set2",
    errorbar=None            
)
plt.title(" Average BMI by Diabetes Status", fontsize=16, fontweight="bold")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Average BMI")
plt.legend([],[], frameon=False)  # hide redundant legend
plt.show()
```

```python
df_balanced = df.copy()
```

```python
df_balanced.loc[df_balanced.sample(frac=0.5, random_state=42).index, "Diabetes_binary"] = 1.0

print(df_balanced["Diabetes_binary"].value_counts())
```

```python
plt.figure(figsize=(8, 6))
sns.barplot(
    data=df_balanced,
    x="Diabetes_binary",
    y="BMI",
    hue="Diabetes_binary",   
    palette="Set2",
    errorbar=None,
    legend=False             
)
plt.title(" Average BMI by Diabetes Status", fontsize=16, fontweight="bold")
plt.xlabel("Diabetes (0 = No, 1 = Yes)")
plt.ylabel("Average BMI")
plt.show()
```

```python
#  Histogram — BMI Distribution
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x="BMI", bins=30, kde=True, color="skyblue")
plt.title(" BMI Distribution with Density Curve", fontsize=16, fontweight="bold")
plt.xlabel("BMI")
plt.ylabel("Frequency")
plt.show()
```

```python
#  Scatter Plot — BMI vs High Blood Pressure
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x="BMI", y="HighBP", hue="Diabetes_binary", 
                palette="Set1", alpha=0.6)
plt.title(" Scatter Plot: BMI vs High Blood Pressure", fontsize=16, fontweight="bold")
plt.xlabel("BMI")
plt.ylabel("High Blood Pressure (0=No, 1=Yes)")
plt.legend(title="Diabetes")
plt.show()
```
